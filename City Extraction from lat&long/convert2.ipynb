```python
import pandas as pd
from time import sleep
import functions as fs
```

```python
df2 = pd.read_csv('df2(10k).csv')
df2
```

```python
l = []
```

```python
for i in range(0,10000, 50):
    new = df2.iloc[i:i+50]
    l.append(new.apply(lambda x: fs.city(x.geo_lat, x.geo_lon),axis =1).values)
    sleep(1.5)
```

```python
len(l)
```

```python
l
```

```python
cities = []
for arr in l:
    for item in arr:
        cities.append(item)
```

```python
len(cities)
```

```python
x = pd.DataFrame(cities, columns=['City'])
x
```

```python
x.to_csv('city2.csv', index=False)
```
