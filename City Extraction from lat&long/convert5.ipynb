```python
import pandas as pd
import functions as fs
from time import sleep
```

```python
df5 = pd.read_csv('df5(5k).csv')
df5
```

```python
l = []
```

```python
for i in range(0,5000, 50):
    new = df5.iloc[i:i+50]
    l.append(new.apply(lambda x: fs.city(x.geo_lat, x.geo_lon),axis =1).values)
    sleep(1.5)
```

```python
len(l)
```

```python
l
```

```python
cities = []
for arr in l:
    for item in arr:
        cities.append(item)
```

```python
len(cities)
```

```python
x = pd.DataFrame(cities, columns=['City'])
x
```

```python
x.to_csv('city5.csv',index=False)
```
