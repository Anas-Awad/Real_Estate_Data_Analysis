```python
import pandas as pd
import functions as fs
from time import sleep
```

```python
df3 = pd.read_csv('df3(10k).csv')
df3
```

```python
l = []
```

```python
for i in range(0,10000, 50):
    new = df3.iloc[i:i+50]
    l.append(new.apply(lambda x: fs.city(x.geo_lat, x.geo_lon),axis =1).values)
    sleep(1.5)
```

```python
len(l)
```

```python
l
```

```python
cities = []
for arr in l:
    for item in arr:
        cities.append(item)
```

```python
len(cities)
```

```python
x = pd.DataFrame(cities,columns=['City'])
x
```

```python
x.to_csv('city3.csv',index=False)
```
