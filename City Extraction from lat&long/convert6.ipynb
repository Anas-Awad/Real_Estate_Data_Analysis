```python
import pandas as pd
from time import sleep
import functions as fs
```

```python
df6 = pd.read_csv('df6(2490).csv')
df6
```

```python
l = []
```

```python
for i in range(0,2490,30):
    new = df6.iloc[i:i+30]
    l.append(new.apply(lambda x: fs.city(x.geo_lat, x.geo_lon),axis =1).values)
    sleep(1.5)
```

```python
len(l)
```

```python
l
```

```python
cities = []
for arr in l:
    for item in arr:
        cities.append(item)
```

```python
len(cities)
```

```python
x = pd.DataFrame(cities, columns= ['City'])
x
```

```python
x.to_csv('city6.csv', index=False)
```
